```python
# viewing and understanding the data
import pandas as pd
df= pd.read_csv("melb_data.csv")
df.isnull().sum()
df.info()
df.describe()
```

```python
#cleaning headers and labels
df.rename(columns={
'lattitude': 'latitude',
    'longtitude': 'longitude',
    'landsize': 'land_size',
    'buildingarea': 'building_area',
    'yearbuilt': 'year_built',
    'councilarea': 'council_area',
    'regionname': 'region_name',
    'propertycount': 'property_count'
}, inplace=True)
df.columns
df.columns = [col.lower() for col in df.columns]
df.columns.to_list()
```

```python
#handling missing values
df.dropna(subset=["car"], inplace= True)
df.car.isnull().sum()
df['building_area'] = df['building_area'].fillna(df['building_area'].median())
df['building_area'].isnull().sum()
df = df.dropna(subset=['year_built'])
df = df.copy() # Explicitly make a copy to avoid SettingWithCopyWarning
df["council_area"] = df["council_area"].fillna("Other")
```

```python
#handling data types
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y')
df['year_sold'] = df['date'].dt.year
df['month_sold'] = df['date'].dt.month
# Displaying the first few rows with the new columns for verification
print(df[['date', 'year_sold', 'month_sold']].head())
```

```python
#Categorical columns (strings everywhere)
df= pd.get_dummies(df, columns=[
   'type', 'method', 'region_name'
])
```

```python
#logical categories cleaning
df['postcode'] = df['postcode'].astype(str)
df= df.drop(columns=['address'])
```

```python
#final cleaning. making it ready for ML
df = df.drop(columns=[
    'suburb',
    'sellerg',
    'council_area',
    'postcode',
    'date'
], errors='ignore')
df = df.drop(columns=['type_u', 'method_PI', 'region_name_Western Victoria'], errors='ignore')
df.info()
```
