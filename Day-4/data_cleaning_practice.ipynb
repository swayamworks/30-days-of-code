```python
import pandas as pd
df= pd.read_csv("marketing_campaign_data_messy.csv")
```

```python
#Headers Cleaning
df.columns.to_list()
df.columns = df.columns.str.strip(" ").str.lower().str.replace(" ", "_")
df.columns.to_list()
```

```python
#type conversion and currency cleaning
spend= df['spend'].astype(str).str.contains(r"\$")
print(df.loc[spend, ['campaign_id', 'spend']])
df["spend1"]= df['spend'].astype(str).str.replace(r'[^\d.\-]', '', regex= True)
df['spend1'] = df['spend1'].astype(float)
print(df.loc[spend, ['campaign_id', 'spend1']])
```

```python
#Categorical Typos
import numpy as np
print(df['channel'].unique())
cleanup_map=  {
    'Facebok': "Facebook",
    'Insta_gram': 'Instagram',
    'Tik_Tok': 'TikTok',
    'E-mail': 'Email',
    'N/A': np.nan
}
df['channel']= df['channel'].replace(cleanup_map)
df
```

```python
#handling booleans
print(df['active'].unique())
new_map= {
    'Yes': True,
    'No': False,
    1: True,
    0: False,
    'Y': True,
    'N': False
}
df['active']= df['active'].replace(new_map)

df
```

```python
#date parsing
print(df['start_date'].dtype)
df['start_date']= pd.to_datetime(df['start_date'], errors='coerce')
df['end_date']= pd.to_datetime(df['end_date'],dayfirst= True, errors='coerce')
df
df['start_date'].dtype
```

```python
df= df.loc[:, ~df.columns.duplicated()]
df
```

```python
#logical integrity (time travel)
time_travel_mask= df['end_date'] < df['start_date']
df.loc[time_travel_mask, ['campaign_id', 'start_date', 'end_date']].head()
df.loc[time_travel_mask, 'end_date']= df.loc[time_travel_mask, 'start_date'] + pd.Timedelta(days=30)
print(df.loc[time_travel_mask, ['campaign_id', 'start_date', 'end_date']].head())
```

```python
#HANDLING OUTLIERS
df['spend1'] = df['spend1'].astype(float)
q1= df['spend1'].quantile(0.25)
q3= df['spend1'].quantile(0.75)
IQR= q3-q1
upper_limit = q3 + 1.5 * IQR
lower_limit = q1 - 1.5 * IQR
outlier_mask= (df['spend1'] > upper_limit) | (df['spend1'] < lower_limit)
print(df.loc[outlier_mask, ['campaign_id', 'spend1']])
```
